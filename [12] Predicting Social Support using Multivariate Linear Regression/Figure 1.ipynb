```python
## Data frame 
import pandas as pd

import numpy as np
from patsy import dmatrices

from statsmodels.compat import lzip
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.regression.linear_model import OLSResults

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rc("figure", figsize=(16, 8))
plt.rc("font", size=14)
```

```python
data_main= {'Country Name':["Canada", "Canada",
                       "Egypt", "Egypt",
                       "Germany", "Germany",
                       "Kenya", "Kenya"], 
            'year':[2015, 2016,
                   2015, 2016,
                   2015, 2016,
                   2015, 2016],
           'Social_support':[0.97, 0.96,
                             0.86, 0.87,
                             0.91, 0.93,
                             0.85, 0.85]}
 
df_main = pd.DataFrame(data_main)

df_main
```

```python
data_edu = {'Country Name':["Canada", 
                       "Egypt", 
                       "Germany",
                       "Kenya",
                      "Papua New Guinea"], 
            2015: [0.67, 0.54, 0.78,0.45, 0.48],
            2016: [0.87, 0.55, 0.80, 0.46, 0.49],
            2017: [0.90, 0.54, 0.82, 0.45, 0.50] }
 
df_edu = pd.DataFrame(data_edu)

df_edu
```

```python
data_edu_copy = df_edu
```

```python
data_edu_copy.set_index(['Country Name'], inplace=True)

data_edu_copy.head(5)
```

```python
data_edu_copy.columns = pd.MultiIndex.from_product([['Graduation_Ratio'], data_edu_copy.columns])
```

```python
data_edu_stacked = data_edu_copy.stack(level=1)
data_edu_stacked.head(5)
```

```python
data_edu_stacked.reset_index(inplace=True)
data_edu_stacked
```

```python
data_edu_stacked = data_edu_stacked.rename(columns = {"level_1":"year" })
data_edu_stacked.head(10)
```

```python
main_data_merged = pd.merge(left = df_main, 
                            right = data_edu_stacked, 
                            how = 'left', 
                            on = ["Country Name", "year"])

main_data_merged
```
